```python
import os
from pathlib import Path

from breeze_connect import BreezeConnect
from dotenv import load_dotenv

load_dotenv()
```

```python
breeze = BreezeConnect(api_key=breeze_api_key)
breeze.generate_session(api_secret=breeze_api_secret, session_token=breeze_session_token)
breeze
```

```python
from datetime import datetime, timedelta

start = datetime(2026, 1, 1, 9, 15)
end = datetime(2026, 1, 10, 15, 30)

delta = timedelta(minutes=1000)  # for 1-min data (1000 candles)

current = start
all_data = []

while current < end:
    next_time = current + delta
    
    data = breeze.get_historical_data_v2(
        interval="1minute",
        from_date=current.strftime("%Y-%m-%dT%H:%M:%S.000Z"),
        to_date=min(next_time, end).strftime("%Y-%m-%dT%H:%M:%S.000Z"),
        stock_code="RELIND",
        exchange_code="NSE",
        product_type="cash"
    )
    
    all_data.extend(data["Success"])
    current = next_time
```
